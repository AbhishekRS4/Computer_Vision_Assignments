```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_pixels_shear = np.array(
    [25, 50, 100, 200, 300, 400]
)
```

```python
num_keypoint_matches = np.array(
    [711, 602, 262, 11, 4, 0]
)
```

```python
fig = plt.figure(figsize=(12, 12))
plt.plot(num_pixels_shear, num_keypoint_matches, "o-")
plt.grid()
plt.xlabel("Number of pixels shear applied", fontsize=20)
plt.ylabel("Number of key point matches detected", fontsize=20)
plt.title("Variation of number of key point matches vs number of pixel shear applied", fontsize=20)
fig.show()
```
